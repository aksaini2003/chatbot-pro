# Production Chatbot Application

A production-ready web application built with React + FastAPI featuring a ChatGPT-like interface with LangGraph backend and RAG capabilities.

## Features

### Backend (FastAPI)
- JWT authentication system
- User management with registration/login
- Conversation persistence with SQLite
- PDF document upload and indexing
- LangGraph integration with multiple tools:
  - PDF RAG (Retrieval Augmented Generation)
  - Web search (DuckDuckGo)
  - Calculator
  - Stock price lookup
  - Weather information
  - Currency conversion
- Streaming chat responses
- RESTful API endpoints

### Frontend (React + TypeScript)
- Modern ChatGPT-like interface
- Dark/Light theme support
- Responsive design
- Conversation history sidebar
- PDF upload functionality
- Real-time typing indicators
- Tool execution indicators
- Authentication pages (Login/Register)

## Setup Instructions

### Prerequisites
- Python 3.8+
- Node.js 16+
- npm or yarn

### Backend Setup

1. Navigate to the backend directory:
```bash
cd backend
```

2. Create a virtual environment:
```bash
python -m venv venv
```

3. Activate the virtual environment:
```bash
# Windows
venv\Scripts\activate

# macOS/Linux
source venv/bin/activate
```

4. Install dependencies:
```bash
pip install -r requirements.txt
```

5. Set up environment variables:
- Copy the `.env` file and update the API keys:
  - `GROQ_API_KEY`: Your Groq API key
  - `GOOGLE_API_KEY`: Your Google AI API key
  - `WEATHER_API_KEY`: Your WeatherAPI key
  - `SECRET_KEY`: Generate a secure JWT secret key

6. Start the backend server:
```bash
python main.py
```

The backend will be available at `http://localhost:8000`

### Frontend Setup

1. Navigate to the frontend directory:
```bash
cd frontend
```

2. Install dependencies:
```bash
npm install
```

3. Start the development server:
```bash
npm start
```

The frontend will be available at `http://localhost:3000`

## API Endpoints

### Authentication
- `POST /auth/register` - Register a new user
- `POST /auth/token` - Login user (OAuth2 password flow)
- `GET /auth/me` - Get current user info

### Conversations
- `GET /conversations` - Get user's conversations
- `POST /conversations` - Create new conversation
- `GET /conversations/{id}` - Get conversation details
- `DELETE /conversations/{id}` - Delete conversation
- `POST /conversations/{id}/chat` - Send message (streaming)
- `POST /conversations/{id}/upload` - Upload PDF document

## Usage

1. **Register/Login**: Create an account or login with existing credentials
2. **Create Conversation**: Start a new chat session
3. **Upload PDF**: Upload PDF documents for RAG capabilities
4. **Chat**: Send messages and interact with the AI assistant
5. **Switch Themes**: Toggle between dark and light modes
6. **Manage Conversations**: View history, delete old conversations


### Frontend
Create `.env` file in frontend directory:
```
REACT_APP_API_URL=http://localhost:8000
```

## Technology Stack

### Backend
- **FastAPI**: Modern Python web framework
- **SQLAlchemy**: ORM for database operations
- **JWT**: Authentication tokens
- **LangGraph**: Conversational AI framework
- **FAISS**: Vector database for document retrieval
- **Groq**: LLM provider
- **Google AI**: Embeddings provider

### Frontend
- **React 18**: UI framework
- **TypeScript**: Type safety
- **Tailwind CSS**: Utility-first CSS framework
- **Lucide React**: Icon library
- **Axios**: HTTP client
- **React Router**: Navigation

## Development Notes

- The application uses SQLite for development (can be upgraded to PostgreSQL for production)
- All existing functionality from the original Streamlit app has been preserved
- The frontend includes proper error handling and loading states
- Streaming responses provide real-time chat experience
- Tool execution is shown with visual indicators

## Production Deployment

For production deployment:

1. **Backend**:
   - Use a production database (PostgreSQL/MySQL)
   - Set up proper CORS origins
   - Use environment-specific configuration
   - Deploy with Docker or cloud service

   

2. **Frontend**:
   - Build the production bundle: `npm run build`
   - Serve with a web server (Nginx, Apache, or Vercel)
   - Configure environment variables for production API URL
